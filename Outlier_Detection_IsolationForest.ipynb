```
# Import the machine learning libraries.
import numpy as np
import matplotlib.pyplot as plt
```

```
from sklearn.ensemble import IsolationForest

np.random.seed(6)

random_data = np.random.randn(200,2)  * 20 + 20

clf = IsolationForest(random_state=0)
preds = clf.fit_predict(random_data)
```

```
preds
```

```
random_data.shape
```

```
plt.figure(figsize=(6, 6))
plt.scatter(random_data[:,0], random_data[:,1], c='k')
plt.rc('font', size=18) 
plt.show()
```

```
plt.figure(figsize=(6, 6))
for i in set(preds):
    if i == -1: 
        #outlier according to dbscan
        dbscan_outlier = plt.scatter(random_data[preds==i, 0], random_data[preds==i, 1], c='r')
    else:
        dbscan_inlier = plt.scatter(random_data[preds==i, 0], random_data[preds==i, 1], c='b')
plt.legend((dbscan_outlier,dbscan_inlier),('IsolationForest Outlier', 'IsolationForest Inlier'),scatterpoints=1, loc='lower right', ncol=3, fontsize=12)
plt.show()
```
